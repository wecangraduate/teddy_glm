```python
from peft import PeftModel, PeftConfig, get_peft_config, get_peft_model, LoraConfig, TaskType
from transformers import AutoModel, AutoTokenizer
from accelerate import infer_auto_device_map, init_empty_weights
```

```python
from accelerate import init_empty_weights
from transformers import AutoConfig, AutoModelForCausalLM

model = AutoModel.from_pretrained("THUDM/chatglm-6b", load_in_8bit=False, trust_remote_code=True, device_map='auto').half()
device = torch.device("cuda")
model.to(device)
```

```python
adapters_name = "teddy0413/glm0924_50000"
m = PeftModel.from_pretrained(model, adapters_name)
```

```python
m.to(device)##peft_model_id = "teddy0413/chatglm0506ora"
##config = PeftConfig.from_pretrained(peft_model_id)
```

```python
tokenizer = AutoTokenizer.from_pretrained("THUDM/chatglm-6b", trust_remote_code=True)
```

```python
input_text = '股份有限公司可以由外國人當董監事嗎？'
input_ids = tokenizer.encode(input_text, return_tensors='pt').cuda()
```

```python
input_ids.is_cuda
```

```python
out = m.generate(
    input_ids=input_ids,
    max_length=500,
    temperature=0.001,
    repetition_penalty=1.25


)
answer = tokenizer.decode(out[0])
print(answer)
```
